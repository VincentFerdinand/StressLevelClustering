# Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

selected_features = ['anxiety_level', 'depression', 'sleep_quality', 
                     'headache', 'study_load', 'living_conditions', 
                     'safety', 'academic_performance']
X = stress[selected_features].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

silhouette_scores = []
for n_clusters in range(2, 11):
    agglomerative = AgglomerativeClustering(n_clusters=n_clusters)
    clusters = agglomerative.fit_predict(X_scaled)
    silhouette_scores.append(silhouette_score(X_scaled, clusters))

plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 11))

best_n_clusters = silhouette_scores.index(max(silhouette_scores)) + 2
plt.axvline(x=best_n_clusters, color='red', linestyle='--', label=f'Best n_clusters = {best_n_clusters}')
plt.legend()
plt.show()

agglomerative = AgglomerativeClustering(n_clusters=best_n_clusters)
clusters = agglomerative.fit_predict(X_scaled)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

unique_clusters, counts = np.unique(clusters, return_counts=True)
plt.figure(figsize=(10, 8))
for cluster, count in zip(unique_clusters, counts):
    plt.scatter(X_pca[clusters == cluster, 0], X_pca[clusters == cluster, 1], 
                label=f'Cluster {cluster} ({count} points)')

plt.legend()
plt.title('Agglomerative Clustering of Stress Level')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

# CLASSIFICATION

```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, f1_score, adjusted_rand_score
import matplotlib.pyplot as plt
import itertools

selected_features = ['anxiety_level', 'depression', 'sleep_quality', 
                     'headache', 'study_load', 'living_conditions', 
                     'safety', 'academic_performance']

X = stress[selected_features].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(eps=1.5, min_samples=3)
clusters = dbscan.fit_predict(X_scaled)

stress['cluster'] = clusters

X_classification = stress[selected_features].values
y_classification = clusters
```

```python
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion Matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# K-Folds cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

precision_scores = []
f1_scores = []
accuracy_scores = []
adjusted_rand_scores = []

for train_index, test_index in kf.split(X_classification):
    X_train, X_test = X_classification[train_index], X_classification[test_index]
    y_train, y_test = y_classification[train_index], y_classification[test_index]

    clf = RandomForestClassifier(random_state=42)
    clf.fit(X_train, y_train)

    y_pred = clf.predict(X_test)

    cm = confusion_matrix(y_test, y_pred)

    precision = precision_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')
    accuracy = accuracy_score(y_test, y_pred)
    adjusted_rand = adjusted_rand_score(y_test, y_pred)

    precision_scores.append(precision)
    f1_scores.append(f1)
    accuracy_scores.append(accuracy)
    adjusted_rand_scores.append(adjusted_rand)

    print("\nConfusion Matrix:")
    print(cm)
    plt.figure()
    plot_confusion_matrix(cm, classes=np.unique(y_test), title='Confusion Matrix')
    plt.show()

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

avg_precision = np.mean(precision_scores)
avg_f1 = np.mean(f1_scores)
avg_accuracy = np.mean(accuracy_scores)
avg_adjusted_rand = np.mean(adjusted_rand_scores)

print(f"\nAverage Precision: {avg_precision:.4f}")
print(f"Average F1-Score: {avg_f1:.4f}")
print(f"Average Accuracy Score: {avg_accuracy:.4f}")
print(f"Average Adjusted Rand Score: {avg_adjusted_rand:.4f}")
```
