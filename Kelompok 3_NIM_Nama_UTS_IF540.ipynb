# Density-Based Clustering Algorithm (DBSCAN)

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

selected_features = ['anxiety_level', 'depression', 'sleep_quality', 
                     'headache', 'study_load', 'living_conditions', 
                     'safety', 'academic_performance']

X = stress[selected_features].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

eps_values = [0.1, 0.5, 1.0, 1.5]

best_silhouette_score = -1
best_eps = None
best_min_samples = None

for eps in [0.1, 0.5, 1.0, 1.5, 2.0]:
    for min_samples in range(2, 11):
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        clusters = dbscan.fit_predict(X_scaled)
        
        if len(set(clusters)) > 1:
            silhouette = silhouette_score(X_scaled, clusters)
            
            if silhouette > best_silhouette_score:
                best_silhouette_score = silhouette
                best_eps = eps
                best_min_samples = min_samples

print("Best eps:", best_eps)
print("Best min_samples:", best_min_samples)
print("Best silhouette score:", best_silhouette_score)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Data preparation
selected_features = ['anxiety_level', 'depression', 'sleep_quality', 
                     'headache', 'study_load', 'living_conditions', 
                     'safety', 'academic_performance']
X = stress[selected_features].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Silhouette method to find optimal eps and min_samples
best_silhouette_score = -1
best_eps = None
best_min_samples = None

for eps in [0.1, 0.5, 1.0, 1.5, 2.0]:
    for min_samples in range(2, 11):
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        clusters = dbscan.fit_predict(X_scaled)
        
        if len(set(clusters)) > 1:
            silhouette = silhouette_score(X_scaled, clusters)
            
            if silhouette > best_silhouette_score:
                best_silhouette_score = silhouette
                best_eps = eps
                best_min_samples = min_samples

print("Best eps:", best_eps)
print("Best min_samples:", best_min_samples)
print("Best silhouette score:", best_silhouette_score)

# Clustering with best parameters
dbscan = DBSCAN(eps=best_eps, min_samples=best_min_samples)
clusters = dbscan.fit_predict(X_scaled)

print("Label Cluster:", clusters)

# PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plotting the clusters
unique_clusters, counts = np.unique(clusters, return_counts=True)
plt.figure(figsize=(10, 8))
for cluster, count in zip(unique_clusters, counts):
    if cluster == -1:
        plt.scatter(X_pca[clusters == cluster, 0], X_pca[clusters == cluster, 1], 
                    label=f'Noise ({count} points)', color='gray', alpha=0.3)
    else:
        plt.scatter(X_pca[clusters == cluster, 0], X_pca[clusters == cluster, 1], 
                    label=f'Cluster {cluster} ({count} points)', alpha=0.5)

plt.legend()
plt.title('DBSCAN Clustering of Stress Level')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

#Clustering with DBSCAN
dbscan = DBSCAN(eps=best_eps, min_samples=best_min_samples)
clusters_dbscan = dbscan.fit_predict(X_scaled)

silhouette_avg_dbscan = silhouette_score(X_scaled, clusters_dbscan)
print("DBSCAN Silhouette Score:", silhouette_avg_dbscan)

# Davies-Bouldin Index Score
dbi_score_dbscan = davies_bouldin_score(X_scaled, clusters_dbscan)
print("DBSCAN Davies-Bouldin Index Score:", dbi_score_dbscan)

# Calinski-Harabasz Index Score
chi_score_dbscan = calinski_harabasz_score(X_scaled, clusters_dbscan)
print("DBSCAN Calinski-Harabasz Index Score:", chi_score_dbscan)
```

# Mean Shift Clustering

```python
from sklearn.cluster import MeanShift
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Data preparation
selected_features = ['anxiety_level', 'depression', 'sleep_quality', 
                     'headache', 'study_load', 'living_conditions', 
                     'safety', 'academic_performance']
X = stress[selected_features].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Clustering with Mean Shift
meanshift = MeanShift()
clusters = meanshift.fit_predict(X_scaled)

print("Number of clusters:", len(set(clusters)))

# PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plotting the clusters
unique_clusters, counts = np.unique(clusters, return_counts=True)
plt.figure(figsize=(10, 8))
for cluster, count in zip(unique_clusters, counts):
    plt.scatter(X_pca[clusters == cluster, 0], X_pca[clusters == cluster, 1], 
                label=f'Cluster {cluster} ({count} points)')

plt.legend()
plt.title('Mean Shift Clustering of Stress Level')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Clustering with Mean Shift
meanshift = MeanShift()
clusters_meanshift = meanshift.fit_predict(X_scaled)

# Calculate silhouette score for Mean Shift
silhouette_avg_meanshift = silhouette_score(X_scaled, clusters_meanshift)
print("Mean Shift Silhouette Score:", silhouette_avg_meanshift)

# Davies-Bouldin Index Score
dbi_score_meanshift = davies_bouldin_score(X_scaled, clusters_meanshift)
print("Mean Shift Davies-Bouldin Index Score:", dbi_score_meanshift)

# Calinski-Harabasz Index Score
chi_score_meanshift = calinski_harabasz_score(X_scaled, clusters_meanshift)
print("Mean Shift Calinski-Harabasz Index Score:", chi_score_meanshift)
```

# Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# Data preparation
selected_features = ['anxiety_level', 'depression', 'sleep_quality', 
                     'headache', 'study_load', 'living_conditions', 
                     'safety', 'academic_performance']
X = stress[selected_features].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Silhouette method to find optimal number of clusters
silhouette_scores = []
for n_clusters in range(2, 11):
    agglomerative = AgglomerativeClustering(n_clusters=n_clusters)
    clusters = agglomerative.fit_predict(X_scaled)
    silhouette_scores.append(silhouette_score(X_scaled, clusters))

# Plot Silhouette scores
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 11))

best_n_clusters = silhouette_scores.index(max(silhouette_scores)) + 2
plt.axvline(x=best_n_clusters, color='red', linestyle='--', label=f'Best n_clusters = {best_n_clusters}')
plt.legend()
plt.show()

# Clustering with best number of clusters
agglomerative = AgglomerativeClustering(n_clusters=best_n_clusters)
clusters = agglomerative.fit_predict(X_scaled)

# PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plotting the clusters
unique_clusters, counts = np.unique(clusters, return_counts=True)
plt.figure(figsize=(10, 8))
for cluster, count in zip(unique_clusters, counts):
    plt.scatter(X_pca[clusters == cluster, 0], X_pca[clusters == cluster, 1], 
                label=f'Cluster {cluster} ({count} points)')

plt.legend()
plt.title('Agglomerative Clustering of Stress Level')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
